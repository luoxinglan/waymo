```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 设置matplotlib使用英文标签
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']

# 读取Excel文件中的所有sheet
file_path = '软件工程.xlsx'
sheets_dict = pd.read_excel(file_path, sheet_name=['2025', '2024', '2023'])

# 初始化一个字典来存储每年的过滤后的数据
filtered_data_by_year = {}

for year, df in sheets_dict.items():
    # 过滤出报考专业名称为“软件工程”的学生数据
    software_engineering_df = df[df['报考专业名称'] == '软件工程']
    
    # 进一步过滤出备注为“中关村校区”、“信创学院”或“唐山”的学生数据
    filtered_df = software_engineering_df[software_engineering_df['备注'].isin(['Zhongguancun Campus', 'Xinchuang College', 'Tangshan'])]
    
    # 将复试成绩换算成500分制
    filtered_df['Interview results (500 scale)'] = filtered_df['Interview results'] * 5
    
    # 存储每年的过滤后的数据
    filtered_data_by_year[year] = filtered_df
    
    # 打印每年过滤后的数据
    print(f"Filtered data for {year}:")
    print(filtered_df)
    
    # 绘制成绩分布的热力图
    plt.figure(figsize=(10, 6))
    correlation_matrix = filtered_df[['Total score of preliminary examination', 'Interview results (500 scale)', 'Total score']].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title(f'Correlation Heatmap of Scores and Remarks for Software Engineering Students in {year}')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.show()
    
    # 分组统计不同校区的成绩平均值
    grouped_stats = filtered_df.groupby('备注')[['Total score of preliminary examination', 'Interview results (500 scale)', 'Total score']].mean()
    print(f"Grouped statistics for {year}:")
    print(grouped_stats)

# 初始化一个字典来存储每年每个校区的成绩平均值
average_scores_by_year_and_campus = {}

for year, df in filtered_data_by_year.items():
    grouped_stats = df.groupby('备注')[['Total score of preliminary examination', 'Interview results (500 scale)', 'Total score']].mean()
    average_scores_by_year_and_campus[year] = grouped_stats

# 转换为长格式以便绘图
long_format_data = []
for year, stats in average_scores_by_year_and_campus.items():
    for campus, row in stats.iterrows():
        long_format_data.append({
            'Year': int(year),
            'Campus': campus,
            'Total score of preliminary examination': row['Total score of preliminary examination'],
            'Interview results (500 scale)': row['Interview results (500 scale)'],
            'Total score': row['Total score']
        })

long_format_df = pd.DataFrame(long_format_data)

# 绘制三项分数的三年走势折线图，并标明数据点的具体数值
plt.figure(figsize=(14, 10))

# 初试总分
plt.subplot(3, 1, 1)
for campus in long_format_df['Campus'].unique():
    subset = long_format_df[long_format_df['Campus'] == campus]
    plt.plot(subset['Year'], subset['Total score of preliminary examination'], marker='o', label=campus)
    for i, txt in enumerate(subset['Total score of preliminary examination']):
        plt.annotate(f'{txt:.2f}', (subset['Year'].iloc[i], txt), textcoords="offset points", xytext=(0,10), ha='center')
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Preliminary Examination Score Trend by Campus (2023-2025)')
plt.legend()
plt.grid(True)

# 复试成绩（500分制）
plt.subplot(3, 1, 2)
for campus in long_format_df['Campus'].unique():
    subset = long_format_df[long_format_df['Campus'] == campus]
    plt.plot(subset['Year'], subset['Interview results (500 scale)'], marker='o', label=campus)
    for i, txt in enumerate(subset['Interview results (500 scale)']):
        plt.annotate(f'{txt:.2f}', (subset['Year'].iloc[i], txt), textcoords="offset points", xytext=(0,10), ha='center')
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Interview Results (500 Scale) Trend by Campus (2023-2025)')
plt.legend()
plt.grid(True)

# 总成绩
plt.subplot(3, 1, 3)
for campus in long_format_df['Campus'].unique():
    subset = long_format_df[long_format_df['Campus'] == campus]
    plt.plot(subset['Year'], subset['Total score'], marker='o', label=campus)
    for i, txt in enumerate(subset['Total score']):
        plt.annotate(f'{txt:.2f}', (subset['Year'].iloc[i], txt), textcoords="offset points", xytext=(0,10), ha='center')
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Total Score Trend by Campus (2023-2025)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()



```
